# 1. Introduction
## Data Visualization Principles

A well-designed visualization turns raw data into actionable insight. This notebook teaches essential principles of effective chart design using Matplotlib and Seaborn.


## 2. Import Libraries & Create Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# Sample dataset: Monthly website visits by source
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Organic": [320, 450, 500, 480, 530, 610],
    "Paid": [200, 220, 210, 190, 230, 250],
    "Referral": [120, 130, 125, 140, 150, 160],
    "Social": [80, 90, 100, 110, 120, 130]
}
df = pd.DataFrame(data)
df
```

## 3. Principle 1: Clarity over Complexity


Good visualizations should clearly show the key message. Avoid 3D effects, excessive gridlines, or confusing colors.

### Bad Example


```python
plt.plot(df["Month"], df["Organic"], linestyle='--', marker='o', label="Organic")
plt.plot(df["Month"], df["Paid"], linestyle=':', marker='x', label="Paid")
plt.plot(df["Month"], df["Referral"], linestyle='-', marker='s', label="Referral")
plt.plot(df["Month"], df["Social"], linestyle='-.', marker='d', label="Social")
plt.title("Website Traffic")
plt.legend()
plt.show()
```

### Good Example

```python
df_melted = df.melt(id_vars='Month', var_name='Source', value_name='Visits')
plt.figure(figsize=(10,6))
sns.lineplot(data=df_melted, x='Month', y='Visits', hue='Source', marker='o')
plt.title("Monthly Website Visits by Source", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Number of Visits")
plt.tight_layout()
plt.show()
```

## 4. Principle 2: Choose the Right Chart Type


Use
- **Line plots** for trends over time
- **Bar charts** for comparisons
- **Scatter plots** for relationships

### Bar Chart Example: May Data Comparison

```python
may_data = df[df["Month"] == "May"].drop("Month", axis=1).T
may_data.columns = ["Visits"]
may_data.plot(kind="bar", legend=False, color="skyblue")
plt.title("Traffic Sources in May")
plt.ylabel("Visits")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## 5. Principle 3: Label Everything

- Use axis labels, units, titles
- Avoid relying only on color

```python
plt.figure(figsize=(6, 4))
sns.barplot(x=["Organic", "Paid", "Referral", "Social"], y=df[df["Month"] == "Jun"].iloc[0,1:])
plt.title("June Website Traffic by Source")
plt.xlabel("Source")
plt.ylabel("Visits")
plt.tight_layout()
plt.show()
```

## Summary of Best Practices

- Keep it simple and clear  
- Pick the right chart for the message  
- Label your axes and chart title  
- Avoid unnecessary effects  
- Use consistent, accessible colors
