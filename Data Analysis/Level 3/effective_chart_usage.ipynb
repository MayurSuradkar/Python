# Effective Chart Usage

Choosing the right chart is critical in telling your data story. This notebook guides you through the most common chart types and when to use them.

## Goal
Learn when to use which chart and how to choose the most effective one for your data story.

## Import Libraries & Sample Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample dataset
data = {
    'Product': ['A', 'B', 'C', 'D'],
    'Sales': [120, 300, 150, 90],
    'Profit': [30, 80, 40, 10],
    'Region': ['East', 'West', 'East', 'South']
}
df = pd.DataFrame(data)
df
```

## Bar Chart
_Compare Categories_

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Product', y='Sales', data=df)
plt.title("Sales per Product")
plt.show()
```

> **Use When**: Comparing quantities across discrete categories (e.g., products, regions).


## Pie Chart

_Proportion of Whole (Use with Caution)_


```python
plt.figure(figsize=(5,5))
plt.pie(df['Sales'], labels=df['Product'], autopct='%1.1f%%', startangle=90)
plt.title("Sales Distribution")
plt.axis('equal')
plt.show()
```

> **Use When**: Showing proportions (but avoid if categories are too many or similar in size).

## Line Chart
_Trends Over Time_

```python
# Sample time-series data
time_df = pd.DataFrame({
    'Month': pd.date_range(start='2025-01', periods=6, freq='M'),
    'Revenue': [1000, 1200, 1500, 1400, 1600, 1800]
})

plt.figure(figsize=(7,4))
sns.lineplot(x='Month', y='Revenue', data=time_df, marker='o')
plt.title("Monthly Revenue Trend")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

> **Use When**: Tracking trends, changes, or patterns over continuous time intervals.

## Scatter Plot
_Relationships Between Variables_






```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Sales', y='Profit', hue='Region', data=df, s=100)
plt.title("Sales vs Profit by Region")
plt.show()
```

> **Use When**: Identifying relationships or clusters between two numeric variables.

## Box Plot
_Distribution & Outliers_


```python
plt.figure(figsize=(5,4))
sns.boxplot(y='Profit', data=df)
plt.title("Profit Distribution with Outliers")
plt.show()
```

> **Use When**: Understanding distribution, median, quartiles, and detecting outliers.

## Heatmap
_Correlation or Intensity_

```python
# Correlation heatmap
corr = df[['Sales', 'Profit']].corr()

plt.figure(figsize=(4,3))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

> **Use When**: Showing correlation matrices or intensity of values across 2D grids.

## Chart Decision Guide

| Task | Best Chart Type |
|:------|:-----------------|
| Compare categories | Bar Chart |
| Show part of whole | Pie Chart, Donut Chart |
| Show trends | Line Chart |
| Visualize distribution | Histogram, Box Plot |
| Show relationships | Scatter Plot |
| Display correlation | Heatmap |


## Summary

- Use bar charts for comparison, line charts for trends, and scatter plots for relationships.
- Avoid pie charts for too many categories.
- Always choose a chart that enhances clarity, not just decoration.
