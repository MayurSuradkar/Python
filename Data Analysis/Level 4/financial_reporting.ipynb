# Financial Reporting with Pandas & Matplotlib


## Objective
This notebook demonstrates how to create basic financial reports using Python.  
We'll simulate company financial data (Revenue, Expenses, Profit) and generate  
useful summaries and visualizations.

## Why Financial Reporting Matters
Financial reporting helps businesses track performance, meet regulatory compliance,  
and make strategic decisions. Key goals include:
- Summarizing revenue and expenses
- Tracking trends over time
- Visualizing financial KPIs


## Step 1: Importing a necessary library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Step 2: Create a sample dataset


```python
months = pd.date_range(start="2024-01-01", periods=12, freq="M")

np.random.seed(42)  # reproducible results
revenue = np.random.randint(50000, 100000, size=12)
expenses = np.random.randint(30000, 70000, size=12)
profit = revenue - expenses

df = pd.DataFrame({
    "Month": months,
    "Revenue": revenue,
    "Expenses": expenses,
    "Profit": profit
})

print("Sample Financial Data")
df
```

## Step 3: Basic Summary Statistics



```python
print("\nSummary Statistics:")
df.describe()
```

## Step 4: Profit Margin Calculation


```python
df["Profit Margin (%)"] = (df["Profit"] / df["Revenue"] * 100).round(2)
print("\n Added Profit Margin column:")
df.head()
```

## Step 5: Visualization - Revenue vs Expenses


```python
plt.figure(figsize=(10,5))
plt.plot(df["Month"], df["Revenue"], marker='o', label="Revenue", color="green")
plt.plot(df["Month"], df["Expenses"], marker='o', label="Expenses", color="red")
plt.title("Revenue vs Expenses Over Time")
plt.xlabel("Month")
plt.ylabel("Amount ($)")
plt.legend()
plt.grid(True)
plt.show()
```

## Step 6: Visualization - Profit Trend


```python
plt.figure(figsize=(10,5))
plt.bar(df["Month"].dt.strftime("%b"), df["Profit"], color="skyblue")
plt.title("Monthly Profit")
plt.xlabel("Month")
plt.ylabel("Profit ($)")
plt.axhline(0, color='black', linewidth=1)
plt.show()
```

## Step 7: KPI Summary Report


```python
total_revenue = df["Revenue"].sum()
total_expenses = df["Expenses"].sum()
total_profit = df["Profit"].sum()
avg_profit_margin = df["Profit Margin (%)"].mean()
```

```python
print("\n**KPI Summary**")
print("----------------------")
print(f"Total Revenue: ${total_revenue:,.2f}")
print(f"Total Expenses: ${total_expenses:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
print(f"Average Profit Margin: {avg_profit_margin:.2f}%")
```
