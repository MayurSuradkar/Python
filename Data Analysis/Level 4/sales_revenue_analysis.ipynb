# Sales / Revenue Analysis Project
This notebook demonstrates a basic **sales analysis workflow** including:
- Data overview & cleaning
- Exploratory Data Analysis (EDA)
- Visualizations
- Insights generation


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load sample dataset
df = pd.read_csv('/content/sample_sales_data.csv')
df.head()
```

## 1. Data Overview

```python
df.info()
df.describe()
```

## 2. Sales Trends Over Time

```python
sales_trend = df.groupby('Date')['Revenue'].sum()
plt.figure(figsize=(10,5))
plt.plot(sales_trend.index, sales_trend.values, marker='o')
plt.title('Total Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.show()
```

## 3. Top Products by Revenue

```python
top_products = df.groupby('Product')['Revenue'].sum().sort_values(ascending=False)
sns.barplot(x=top_products.values, y=top_products.index)
plt.title('Top Products by Revenue')
plt.xlabel('Revenue')
plt.ylabel('Product')
plt.show()
```

## 4. Revenue by Category

```python
category_revenue = df.groupby('Category')['Revenue'].sum()
category_revenue.plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))
plt.title('Revenue Share by Category')
plt.ylabel('')
plt.show()
```

## 5. Key Insights
- Identify peak sales periods.
- Highlight top-performing products.
- Detect low-performing categories.
