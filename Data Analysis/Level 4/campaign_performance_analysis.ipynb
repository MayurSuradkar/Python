# Campaign Performance Analysis Notebook

## Goal

This notebook demonstrates how to analyze marketing campaign performance using KPIs like CTR, Conversion Rate, and ROI.

We'll create a sample dataset, compute metrics, and visualize results for better decision-making.

# 1. Introduction
Marketing campaigns often aim to attract customers and drive sales.Analyzing their performance helps businesses optimize strategies and budgets.

## Key metrics include
- Impressions: How many times the ad was shown.
- Clicks: How many times users clicked the ad.
- Conversions: How many people completed a desired action (purchase, signup).
- Cost: Amount spent on the campaign.
- Revenue: Money earned from the campaign.

```python
# Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 2. Create Sample Dataset

```python
data = {
    "Campaign": ["Spring Sale", "Summer Promo", "Back to School", "Holiday Deals", "Clearance Sale"],
    "Impressions": [10000, 15000, 12000, 18000, 9000],
    "Clicks": [500, 750, 600, 900, 450],
    "Conversions": [50, 80, 65, 120, 40],
    "Cost": [1000, 1500, 1200, 2000, 800],
    "Revenue": [3000, 5000, 3500, 8000, 2000]
}

df = pd.DataFrame(data)
df
```

# 3. Compute KPIs


```python
df["CTR (%)"] = (df["Clicks"] / df["Impressions"]) * 100
df["Conversion Rate (%)"] = (df["Conversions"] / df["Clicks"]) * 100
df["ROI"] = ((df["Revenue"] - df["Cost"]) / df["Cost"]) * 100

df.round(2)
```

# 4. Visualize Campaign Performance

```python
plt.figure(figsize=(10,5))
sns.barplot(x="Campaign", y="CTR (%)", data=df)
plt.title("Click-Through Rate by Campaign", fontsize=14)
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10,5))
sns.barplot(x="Campaign", y="ROI", data=df)
plt.title("ROI by Campaign", fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

# 5. Insights
- High CTR does not always mean high ROI (e.g., Summer Promo had good CTR but moderate ROI).
- Holiday Deals generated the highest ROI and revenue.
- Clearance Sale had low performance across most KPIs and may require a strategy change.

This type of analysis helps marketers reallocate budgets and improve targeting for better results.
