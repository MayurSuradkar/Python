# **Correlation & Covariance**

_Correlation and covariance are essential tools for understanding relationships between numerical variables._


## What You'll Learn
- What correlation and covariance mean
- How to compute them in pandas
- How to visualize relationships using a heatmap

## Concepts
- **Correlation**: Measures the strength and direction of a linear relationship between two variables. Ranges between -1 and 1.

- **Covariance**: Measures how two variables vary together. Units are not standardized, unlike correlation.


## Import Libraries

````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
````

## Sample Dataset
Weâ€™ll create a small DataFrame to work with:

````python
data = {
    'Hours_Studied': [1, 2, 3, 4, 5, 6, 7],
    'Exam_Score': [50, 55, 65, 70, 75, 85, 90],
    'Sleep_Hours': [8, 7, 6, 6, 5, 4, 3]
}

df = pd.DataFrame(data)
df
````

### 1. Correlation - .corr()

````python
df.corr()
````

Interpretation:

- `Hours_Studied` and `Exam_Score` should show a strong positive correlation.

- `Sleep_Hours` and `Hours_Studied` should show a negative correlation (studying more â†’ less sleep).


#### Visualizing Correlation


````python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
````

> This heatmap helps spot strong/weak correlations visually.

### 2. Covariance â€“ .cov()



````python
df.cov()
````

- Covariance is not normalized and depends on the units of measurement.

- Large values â‰  strong relationship. Always check correlation for that.

## When to Use

| Use Case                  | Use         |
|:------------------------- |:----------- |
| Normalize strength        | Correlation |
| Understand joint movement | Covariance  |
