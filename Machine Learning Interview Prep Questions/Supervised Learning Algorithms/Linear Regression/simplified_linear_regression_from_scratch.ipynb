# Simple Linear Regression from Scratch (No ML Libraries)

We'll build a linear regression model using only NumPy, step-by-step.

Goal: Predict `y` using a line `y = w*x + b`


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate simple data: y = 3x + 4 + noise
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Visualize
plt.scatter(X, y)
plt.title("Training Data")
plt.xlabel("X")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
# Initialize parameters
w = 0
b = 0

lr = 0.1      # learning rate
epochs = 1000 # number of iterations

n = X.shape[0] # number of samples
```

```python
for i in range(epochs):
    y_pred = w * X + b             # predicted values
    error = y_pred - y             # prediction error

    dw = (2/n) * np.sum(X * error) # gradient for weight
    db = (2/n) * np.sum(error)     # gradient for bias

    w -= lr * dw                  # update weight
    b -= lr * db                  # update bias

    if i % 200 == 0:
        loss = np.mean(error ** 2)
        print(f"Epoch {i}: Loss = {loss:.3f}")
```

```python
# Final prediction
y_pred = w * X + b

plt.scatter(X, y, label='Actual')
plt.plot(X, y_pred, color='red', label='Predicted Line')
plt.legend()
plt.title("Fitted Line")
plt.show()

print(f"Final weight (w): {w:.2f}")
print(f"Final bias (b): {b:.2f}")
```

```python
def r2_score(y, y_pred):
    ss_total = np.sum((y - np.mean(y)) ** 2)
    ss_residual = np.sum((y - y_pred) ** 2)
    return 1 - (ss_residual / ss_total)

print("R² Score:", r2_score(y, y_pred))
```
